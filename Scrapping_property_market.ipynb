```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
import re
```

# for zillow

```python
in_data=pd.read_excel('C:\\Users\\Rideema Malji\\OneDrive\\Desktop\\Others\\Upwork\\we scrapping\\input_src.xlsx',sheet_name='New Reports')
links=in_data['Link']
```

```python
#get values from zillow
#url='https://www.zillow.com/homedetails/3859-Ellisboro-Rd-Stokesdale-NC-27357/83184790_zpid/'
property_address=[]
price=[]
zestimate=[]
month_cost_fetched=[]
for url in links.values[:]:
    try:
        print('processing', url)
        header = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
            'referer': url
        }
        html = requests.get(url=url,headers=header)
        #print(html.status_code)
        if html.status_code==200:
            soup1 = soup(html.content,'lxml')
            header = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
                'referer': url
            }

           # Price = soup.findAll('span',attrs={'class':'Text-c11n-8-84-0__sc-aiai24-0 dpf__sc-1me8eh6-0 hLAJE fzJCbY'})
            #Price2 = soup.findAll('span',attrs={'class':'Text-c11n-8-84-0__sc-aiai24-0 gTVYcr'})
            p_c=0
            price_l1=soup1.findAll('span',attrs={'class':'Text-c11n-8-84-0__sc-aiai24-0 dpf__sc-1me8eh6-0 hLAJE fzJCbY'})
            price_l2=soup1.findAll('strong')
            if len(price_l1)>0:
                final_list_price=price_l1
            else:
                final_list_price=price_l2[5:6]
            for link1 in final_list_price:
                p_c=1
                price.append(link1.get_text())
            if p_c==0:
                price.append('')
            else:
                pass
            p_a=0
            add_l1=soup1.findAll('h1',attrs={'class':'Text-c11n-8-84-0__sc-aiai24-0 fsXIkY'})
            add_l2=soup1.findAll('h1',attrs={'class':'Text-c11n-8-84-0__sc-aiai24-0 qxgaF'})
            if len(add_l1)>0:
                final_list=add_l1
            elif len(add_l2)>0:
                final_list=add_l2
            else:
                final_list=[]
            for link2 in final_list:
                property_address.append(link2.get_text())
                p_a=1
            if p_a==0:
                property_address.append('')
            else:
                pass
            z_c=0
            zest_l1=soup1.findAll('span',attrs={'class':'Text-c11n-8-84-0__sc-aiai24-0 gTVYcr'})
            zest_l2=soup1.findAll('span',attrs={'class':'Text-c11n-8-84-0__sc-aiai24-0 eIEmla'})
            if len(zest_l1)>0:
                final_zest=zest_l1
            elif len(zest_l2)>0:
                final_zest=zest_l2[1:2]
            for link3 in final_zest:
                x = re.search("^\$.*", link3.get_text())
                if x is None:
                    pass
                else:
                    z_c=1
                    zestimate.append(x.string)
            if z_c==0:
                zestimate.append('')
            else:
                pass
            monthly_cost=soup1.findAll('span',attrs={'class':'Text-c11n-8-84-0__sc-aiai24-0 kdSWZE'}) 
            if len(monthly_cost)==0:
                monthly_cost=[]
            else:
                 monthly_cost=soup1.findAll('span',attrs={'class':'sc-AHaJN kqOskK'}) 
            for link4 in monthly_cost:
                month_cost_fetched.append(link4.get_text())
    except:
            print('Error while collecting data from : ',url)

zillow_dataset=pd.DataFrame({'Price':price,'Property_address':property_address,'zestimate':zestimate})
```

# Value Estimator 1 

```python
books = zillow_dataset['Property_address'].values.tolist()
```

```python
import requests
from bs4 import BeautifulSoup as soup
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
options_sel=webdriver.ChromeOptions()
options_sel.add_argument('headless')
d = webdriver.Chrome('/path/to/chromedriver',options=options_sel)
d.implicitly_wait(5)
val_est1=[]
###place the adress to look for here 
for indx,book in enumerate(books):
    if indx==0:
        d.get("https://www.remax.com/home-value-estimates")
        bttn=d.find_element(By.XPATH,"//button[normalize-space()='Accept All Cookies']")
        bttn.click()
        e = d.find_element(By.CLASS_NAME, "search.with-button")
        e.send_keys(book)
        time.sleep(2)
        e.send_keys(Keys.ENTER)
        time.sleep(5)
        q=d.find_element(By.XPATH,"//button[normalize-space()='Continue']")
        q.click()
        time.sleep(3)
        w=d.find_element(By.CSS_SELECTOR,"div[data-test='smartZip-estimate-card'] h4")
        print(w.text)
        get_url = d.current_url
        print("The current url is:"+str(get_url))
        val_est1.append(w.text)
    else:
        try:
            d.get("https://www.remax.com/home-value-estimates")
            e = d.find_element(By.CLASS_NAME, "search.with-button")
            e.send_keys(book)
            time.sleep(2)
            e.send_keys(Keys.ENTER)
            time.sleep(5)
            q=d.find_element(By.XPATH,"//button[normalize-space()='Continue']")
            q.click()
            time.sleep(3)
            w=d.find_element(By.CSS_SELECTOR,"div[data-test='smartZip-estimate-card'] h4")
            print(w.text)
            get_url = d.current_url
            #print("The current url is:"+str(get_url))
            val_est1.append(w.text)
        except:
            print('Error for',book)
            val_est1.append('error')
```

```python
zillow_dataset['value estimator1']=val_est1
```

# Columns S ='Monthly my rent rates'

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
options_sel1=webdriver.ChromeOptions()
options_sel1.add_argument('headless')
r = webdriver.Chrome('/path/to/chromedriver',options=options_sel1)
month_my_rent_rate=[]
for book in books:
    try:
        r.get("https://www.redfin.com/rental-estimate")
        e = r.find_element(By.XPATH, "//div[@class='searchBoxWrapper AvmSearchBox']//input[@id='search-box-input']")
        time.sleep(2)
        e.send_keys(book)
        e.send_keys(Keys.ENTER)
        time.sleep(3)
        monthly_rent=r.find_element(By.CSS_SELECTOR,".predictedRentalRange")
        print('Property:',book,'Monthly_rent:',monthly_rent.text)
        month_my_rent_rate.append(monthly_rent.text)
        get_url = r.current_url
    except:
        print('Error processing:',book)
        month_my_rent_rate.append('error')
```

```python
zillow_dataset['monthly_estimate']=month_my_rent_rate
```

```python
import requests
from bs4 import BeautifulSoup as soup
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
options_sel2=webdriver.ChromeOptions()
options_sel2.add_argument('headless')
boa = webdriver.Chrome('/path/to/chromedriver',options=options_sel2)
boa.implicitly_wait(5)
boa_estimate=[]
for idx,book in enumerate(books):
    try:
        if idx==0:
            boa.get("https://homevaluerealestatecenter.bankofamerica.com/")
            time.sleep(8)
        else:
            boa.get("https://homevaluerealestatecenter.bankofamerica.com/")
            time.sleep(2)
        e = boa.find_element(By.XPATH, "//input[@id='address']")
        e.send_keys(book)
        time.sleep(3)
        e.send_keys(Keys.ARROW_DOWN)
        e.send_keys(Keys.ENTER)
        try:
            chase_value=boa.find_element(By.XPATH,"//div[@class='hvt-property__estimate-value']")
            print(chase_value.text) 
            boa_estimate.append(chase_value.text)
        except:
            print('error') 
            boa_estimate.append('error')
    except:
        print('Error while processing :',chase_value.text)
        continue
```

```python
zillow_dataset['BOA_Estiamtes']=boa_estimate
```
